###  Performing Linear Regression

### Splitting data into training and testing 

It is immportant to split the dataset into training and testing so that model can learn from training dataset and perform on testing dataset. The test size for the dataset varies from industry to industry. However, a standard of 70-30 is taken in general, where 70 % of the data set is used for training and 30 % of the data is used for testing

```python
from sklearn.model_selection import train_test_split 
# scikit learns model selection helps us in splitting the data for training and testing

X_train, X_test, Y_train, Y_test=np.array(train_test_split(LR['X'], LR['Y'], test_size=0.1))

#np.array is used so that numpy can store the info of the training and testing dataset
#Test size is kept only at 10 %, However, this value can be changed as the requirements of the industry.
```

### Building a linear regression model

```python
from sklearn.linear_model import LinearRegression # linear_model from scikit learn is the only way to import the LinearRegression Algorithm

reg = LinearRegression() #for convinenece of making the code easy to write the variable reg is used to initiate LinearRegression

reg.fit(X_train.values.reshape(-1,1), Y_train.values.reshape(-1,1)) 
#using the fit function to add values into the regression algorithm, and .value.reshape(-1,1) is used because,
#algorithm is unable to compute unless we reshape the data into (-1,1)
```

Accuracy from the training dataset

```python
print('Score: ', reg.score(X_train.values.reshape(-1,1), Y_train.values.reshape(-1,1)))
```

Accuracy from the test dataset

```python
print('Score: ', reg.score(X_test.values.reshape(-1,1), Y_test.values.reshape(-1,1)))
```

# Linear Regression with three variables

It is an exaple to show the difference in performing linear regression when more than one dependent variable is involved.

Steps taken to create this linear regression is more or less the same



### Importing data and preprocessing

```python
co2_df=pd.read_csv(r'E:\LYIT\Eoghan Furey ML\Assignment 2 Practical\1-Linear Regression\global_co2.txt')
print(co2_df.head(10))
```

```python
temp_df=pd.read_csv(r'E:\LYIT\Eoghan Furey ML\Assignment 2 Practical\1-Linear Regression\annual_temp.txt')
print(temp_df.head(10))
```

```python
# Clean data

co2_df = co2_df.ix[:,:2] # Keep only total CO2 #.ix is used for slilcing 
co2_df = co2_df.ix[co2_df['Year'] >= 1960]   # Keep data only from year 1960 - 2010
co2_df.columns=['Year','CO2']                # Renaming columns
co2_df = co2_df.reset_index(drop=True)      # Reseting index

temp_df = temp_df[temp_df.Source != 'GISTEMP']                              # Keep only one source
temp_df.drop('Source', inplace=True, axis=1)                                # Drop name of source
temp_df = temp_df.reindex(index=temp_df.index[::-1])                        # Reset index
temp_df = temp_df.ix[temp_df['Year'] >= 1960].ix[temp_df['Year'] <= 2010]   # Keep only 1960 - 2010
temp_df.columns=['Year','Temperature']                                      # Rename columns
temp_df = temp_df.reset_index(drop=True)                                    # Reset index

print(co2_df.head())

print(temp_df.head())

#printinng head of both co2 and temp dataset, by default head(5). This number can be changed as per one's convinence.
```

```python

# Concatenate

# Brining both the datasets together by using concatination function of pandas dataframe.

climate_change_df = pd.concat([co2_df, temp_df.Temperature], axis=1)

print(climate_change_df.head())
```

###  Visuallize in 3D

```python
from mpl_toolkits.mplot3d import Axes3D #mpl_toolkits.mplot3d is a matplotlib library used for 3D plotting
fig=plt.figure()
fig.set_size_inches(12.5, 7.5)

ax = fig.add_subplot(111, projection='3d')

ax.scatter(xs=climate_change_df['Year'], ys=climate_change_df['Temperature'], zs=climate_change_df['CO2'])

ax.set_ylabel('Relative tempature'); ax.set_xlabel('Year'); ax.set_zlabel('CO2 Emissions')
ax.view_init(10, -45)
```


### Projected 2D plots

```python
f, axarr = plt.subplots(2, sharex=True)
f.set_size_inches(12.5, 7.5)
axarr[0].plot(climate_change_df['Year'], climate_change_df['CO2'])
axarr[0].set_ylabel('CO2 Emissions')
axarr[1].plot(climate_change_df['Year'], climate_change_df['Temperature'])
axarr[1].set_xlabel('Year')
axarr[1].set_ylabel('Relative temperature')
```

### Linear Regression

```python
X = climate_change_df.as_matrix(['Year'])
Y = climate_change_df.as_matrix(['CO2', 'Temperature']).astype('float32')

X_train, X_test, y_train, y_test = np.asarray(train_test_split(X, Y, test_size=0.1))
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python

print('Score: ', reg.score(X_test.reshape(-1, 1), y_test))
```

### Plot regression and visualize results

```python
x_line = np.arange(1960,2011).reshape(-1,1)
p = reg.predict(x_line).T
```

### 3D plotting

```python
fig2 = plt.figure()
fig2.set_size_inches(12.5, 7.5)
ax = fig2.add_subplot(111, projection='3d')
ax.scatter(xs=climate_change_df['Year'], ys=climate_change_df['Temperature'], zs=climate_change_df['CO2'])
ax.set_ylabel('Relative tempature'); ax.set_xlabel('Year'); ax.set_zlabel('CO2 Emissions')
ax.plot(xs=x_line, ys=p[1], zs=p[0], color='green')
ax.view_init(10, -45)
```

### 2D plotting

```python

f, axarr = plt.subplots(2, sharex=True)
f.set_size_inches(12.5, 7.5)
axarr[0].plot(climate_change_df['Year'], climate_change_df['CO2'])
axarr[0].plot(x_line, p[0])
axarr[0].set_ylabel('CO2 Emissions')
axarr[1].plot(climate_change_df['Year'], climate_change_df['Temperature'])
axarr[1].plot(x_line, p[1])
axarr[1].set_xlabel('Year')
axarr[1].set_ylabel('Relative temperature')
```

#### Conclusion:
From the above plots it is very clear that there is a strong coorelation between the X("Dependent Variables") and Y("Independent Variables) and any changes made to the independent variables would lead to changes in the dependent variables.
